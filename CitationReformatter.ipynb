# Adding columns to csv-file

```python
# https://stackoverflow.com/questions/11070527/how-to-add-a-new-column-to-a-csv-file
# Input-1: Name of existing CSV filename
# Input-2: Name of output CSV filename (which will have the updated content)
# Input-3: List with header name & column values
# Example:
# new_list = ['test_hdr', 4, 4, 5, 5, 9, 9, 9]
# add_col_to_csv('exists.csv', 'new-output.csv', new_list)

def add_col_to_csv(csvfile, fileout, new_list):
    with open(csvfile, 'r') as read_f, \
        open(fileout, 'w', newline='') as write_f:
        csv_reader = csv.reader(read_f)
        csv_writer = csv.writer(write_f)
        i = 0
        for row in csv_reader:
            row.append(new_list[i])
            csv_writer.writerow(row)
            i += 1 
```

# Quality of life update for working with a batch of csv-files:

```python
for element in ["Sturm", "Lebovitz"]:
    name_input_paper = "2021-" + element + "-anystyle"
    name_input_paper_file = name_input_paper + ".csv"
    name_output_paper_file_temp = name_input_paper + "-tempSemanticScholar.csv"
    name_output_paper_file_temp2 = name_input_paper + "-tempCrossRef.csv"
    name_output_paper_file = name_input_paper + "-citations.csv"

    # Preperation:
    import csv
    new_list_semantic_scholar_citations = ['SemanticScholarCitations']
    new_list_crossref_citations = ['CrossRefCitations']
    new_list_marker = ['Marker']

    # Read csv-file and fill new lists
    with open(name_input_paper_file, mode='r') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        line_count = 0
        for row in csv_reader:
            extractedCitations = row["Extra"].split(" ")
            foundSSCitation = False
            foundCRCitation = False
            for i in range(len(extractedCitations)):    
                if extractedCitations[i] == "citations":
                    if extractedCitations[i+1] == "(Semantic":
                        foundSSCitation = True
                        new_list_semantic_scholar_citations.append(extractedCitations[i-1])
                    if extractedCitations[i+1] == "(Crossref)":
                        foundCRCitation = True
                        new_list_crossref_citations.append(extractedCitations[i-1])
            if foundSSCitation == False:
                new_list_semantic_scholar_citations.append("No Citations Found")
            if foundCRCitation == False:
                new_list_crossref_citations.append("No Citations Found")
            line_count += 1
            # Use line_count to create marker for later identification of reference in paper
            converted_line_counter = "% s" % line_count
            marker = "[" + converted_line_counter +"]"
            new_list_marker.append(marker)

    add_col_to_csv(name_input_paper_file, name_output_paper_file_temp, new_list_semantic_scholar_citations)
    add_col_to_csv(name_output_paper_file_temp, name_output_paper_file_temp2, new_list_crossref_citations)
    add_col_to_csv(name_output_paper_file_temp2, name_output_paper_file, new_list_marker)
```
